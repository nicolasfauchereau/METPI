```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.html.widgets import interact, fixed
from IPython.html import widgets
```

```python
import Image
```

```python
overlay = './islandoutline.png'
```

```python
background = './glbPrecSeaInd1.gif'
```

```python
def adjust(background, overlay, xoffset, yoffset, xscale, yscale): 
    overlay = Image.open(overlay)
    overlay = overlay.convert("RGBA")
    wo, ho = overlay.size
    background = Image.open(background)
    background = background.convert("RGBA")
    wb, hb = background.size
    arrb = np.asarray(background)
    arrb = arrb[yoffset::, xoffset::, :]
    background = Image.fromarray(np.uint8(arrb))
    wb, hb = background.size
    background = background.resize((int(xscale * wb), int(yscale * hb)), Image.ANTIALIAS)
    background.paste(overlay, (0, 0), overlay)
    arrb = np.asarray(background)
    f, ax = plt.subplots(figsize=(12, 10))
    ax.imshow(arrb[0:ho,0:wo,:])
    plt.setp(ax.get_xticklabels(), visible=False)
    plt.setp(ax.get_yticklabels(), visible=False)    
    plt.show(f)
```

```python
interact(adjust, background = fixed(background), overlay = fixed(overlay), \
         xoffset = widgets.IntSliderWidget(min=0,max=500,step=1,value=0),\
         yoffset = widgets.IntSliderWidget(min=0,max=500,step=1,value=0),\
         xscale = widgets.FloatSliderWidget(min=1,max=10,step=0.1,value=1),\
         yscale = widgets.FloatSliderWidget(min=1,max=10,step=0.1,value=1))
```
